# Exploring shapefile 

Data source: INEGI - http://www.beta.inegi.org.mx/app/mapas/

```python
import geopandas as gpd
from pathlib import Path
```

```python
%matplotlib inline
```

```python
raw_files = Path("../data/raw/topografia/conjunto_de_datos/")
```

```python
file_names = [file.name for file in list(raw_files.glob("*.shp"))]

file_names
```

```python
for file in file_names:
    geo_data = gpd.read_file(f"../data/raw/topografia/conjunto_de_datos/{file}")
        
    with open("../data/interim/marco_geoestadistico_log.txt", "w", encoding='utf-8') as log:
        log.write(f"Filename: {file}")
        log.write(f"Number of rows: {geo_data.shape[0]}")
    
    print("#"*40)
    print(f"Filename: {file}")
    print(f"Number of rows: {geo_data.shape[0]:,} \n")
```
