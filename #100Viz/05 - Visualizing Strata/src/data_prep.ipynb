```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/MMSI_2016.csv", encoding = "utf-8", usecols=["ent_ENH", "est_socio_ENH", "P10_2"])

df.head()
```

```python
dff = df.groupby(["ent_ENH", "est_socio_ENH"])["P10_2"].value_counts().to_frame()

dff.columns = ["count"]
dff.reset_index(inplace = True)
dff.columns = ["entidad", "estrato_socioeconomico", "tono_de_color_de_piel", "cuenta"]

dff.head()
```

```python
dff['entidad'].nunique()
```

```python
dff['estrato_socioeconomico'].nunique()
```

```python
estratos = {
    1: "Bajo",
    2: "Medio Bajo",
    3: "Medio Alto",
    4: "Alto"
}
```

```python
domains = range(1,12)
color = ["#322e25", "#3d230c", "#4a382e", "#684d3c", "#7e6352", "#95765a", "#b59a7f", "#dec09a", "#e1b8b2", "#f2d0d1", "#f9ebeb"]

color_scale = list(zip(domains, color))

dict(color_scale)
```

```python
# http://www.inegi.org.mx/geo/contenidos/geoestadistica/CatalogoClaves.aspx
estados = pd.read_excel("estados.xlsx", encoding = "utf-8")

estados.columns = ["entidad", "nom_entidad", "abr_estado"]
estados.head()
```

```python
dff_names = pd.merge(dff, estados, on = "entidad", how = "left")
dff_names.head()
```

```python
dff.shape
```

```python
dff_names['estrato_socioeconomico_nom'] = dff_names["estrato_socioeconomico"].map(estratos)
```

```python
dff_names.to_csv("data_cleaned.csv", encoding = "utf-8", index = False,)
```
