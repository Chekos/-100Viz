```python
counties = alt.topo_feature(data.us_10m.url, 'counties')
unemp_data = data.unemployment.url


chart = alt.Chart(counties).mark_geoshape().encode(
    color='rate:Q'
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(unemp_data, 'id', ['rate'])
).project(
    type='albersUsa'
).properties(
    width=500,
    height=300
)

chart
```
